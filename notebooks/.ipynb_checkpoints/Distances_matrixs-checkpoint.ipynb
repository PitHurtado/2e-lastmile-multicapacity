# Review Data Distances

```python
import pandas as pd
from src.constants import (
    PATH_DATA_DISTANCES_FROM_DC,
    PATH_DATA_DISTANCES_FROM_SATELLITES,
    PATH_DATA_PIXEL,
    PATH_DATA_SATELLITE,
    PATH_ROOT_SCENARIO,
)
from matplotlib import pyplot as plt
```

```python
path_s = '../data/input_matrix_distance_satellite_to_pixels.xlsx'
df_s = pd.read_excel(path_s)

path_d = '../data/input_matrix_distance_dc_to_pixels.xlsx'
df_d = pd.read_excel(path_d)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
for s in df_s.id_satellite.unique():
    filtered = df_s[df_s.id_satellite == s]
    plt.hist(filtered.distance, alpha=0.5, bins=50, label=f"distance - {s}")
    plt.legend()
    
plt.title(f"Hist distance from all satelites")
plt.show()

df_s.groupby('id_satellite')['distance'].describe()
```

```python
plt.hist(df_d.distance)
plt.title(f"HIST distance from DC")
plt.show()
```
