# Costos de Servir

```python
import json
import numpy as np
from src.instance.instance import Instance
from src.model.deterministic import FlexibilityModel
```

```python
instance_to_solve: Instance = Instance(
        id_instance="expected_non_fixed",
        capacity_satellites={"2": 2, "4": 4, "6": 6, "8": 8, "10": 10, "12": 12},
        is_continuous_x=False,
        alpha=0.5,
        beta=0.5,
        type_of_flexibility=2,
        periods=12,
        N=1,
        is_evaluation=False,
    )
scenario = instance_to_solve.get_scenario_expected()
instance_to_solve.scenarios = {"expected": scenario}
pixels = scenario.pixels
satellites = instance_to_solve.satellites
```

```python
cost_from_satellite = scenario.get_cost_serving("satellite")
fleet_size_from_satellite = scenario.get_fleet_size_required("satellite")
```

```python
total_cost_values = [val["total"] for val in cost_from_satellite.values()]
plt.hist(total_cost_values,bins=100)
plt.title("Costo por item de servir de satellite a pixel")
plt.xlabel("costo unitario por item")
plt.show()
print(f"min {np.min(total_cost_values)}")
print(f"median {np.median(total_cost_values)}")
print(f"max {np.max(total_cost_values)}")
print(f"avg {np.mean(total_cost_values)}")
print("----- percentiles-----")
print(f"p5 {np.percentile(total_cost_values,5)}")
print(f"p25 {np.percentile(total_cost_values,25)}")
print(f"p75 {np.percentile(total_cost_values,75)}")
print(f"p95 {np.percentile(total_cost_values,95)}")
print(f"p99 {np.percentile(total_cost_values,99)}")
```

```python
for t in range(12):
    sum_by_period = 0
    for k, pixel in pixels.items():
        sum_by_period += pixel.drop_by_period[t]#pixel.demand_by_period[t]
    print(f"demanda total periodo {t} - demanda {sum_by_period}")
```

```python
total_fleet_size_from_satellite = [val["fleet_size"] for val in fleet_size_from_satellite.values()]
plt.hist(total_fleet_size_from_satellite,bins=100)
plt.title("fleet size required servir de satellite a pixel")
plt.xlabel("cantidad de vehiculos")
plt.show()
print(f"min {np.min(total_fleet_size_from_satellite)}")
print(f"median {np.median(total_fleet_size_from_satellite)}")
print(f"max {np.max(total_fleet_size_from_satellite)}")
print(f"avg {np.mean(total_fleet_size_from_satellite)}")
print("----- percentiles-----")
print(f"p5 {np.percentile(total_fleet_size_from_satellite,5)}")
print(f"p25 {np.percentile(total_fleet_size_from_satellite,25)}")
print(f"p75 {np.percentile(total_fleet_size_from_satellite,75)}")
print(f"p95 {np.percentile(total_fleet_size_from_satellite,95)}")
print(f"p99 {np.percentile(total_fleet_size_from_satellite,99)}")

print("----- requerido -----")
print(f"total {np.sum(total_fleet_size_from_satellite)}")
print(f"total avg period {np.sum(total_fleet_size_from_satellite)/12}")
print(f"total avg period y satellite {np.sum(total_fleet_size_from_satellite)/(9*12)}")
```

```python
value_by_satellite = {}
for s in satellites.keys():
    value_list = []
    for t in range(12):
        sum_value = 0
        for k, pixel in pixels.items():
            sum_value += fleet_size_from_satellite[(s,k,t)]["fleet_size"]
        value_list.append(sum_value)
    value_by_satellite[s] = value_list
```
